# Exploratory Data Analysis - Sports

```python
# import all packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load Dataset
delivery=pd.read_csv("deliveries.csv")
delivery.head()
```

```python
# Load Dataset
match=pd.read_csv("matches.csv")
match.head()
```

```python
# Checking of missing value
delivery.isnull().sum()
```

```python
# last 3 columns has more missing data so not filling dropping the column.
drop_cols=['player_dismissed', 'dismissal_kind' , 'fielder']
delivery.drop(drop_cols,axis=1)
```

```python
# Checking the missing value
match.isnull().sum()
```

```python
# Last 3 columns have more missing data so dropping those columns.
drop_col=['umpire1','umpire2','umpire3']
match.drop(drop_col,axis=1)
```

```python
delivery.describe()
```

```python
match.describe()
```

```python
# Find Correlation
plt.figure(figsize=(8,8))
sns.heatmap(delivery.corr(),annot=True, cmap="coolwarm")
```

```python
# Find Correlation
plt.figure(figsize=(8,8))
sns.heatmap(match.corr(),annot=True, cmap="coolwarm")
```

```python
# Visualization 1
# Team wins maximum number in IPL.

match['winner'].value_counts().plot(kind="bar",figsize=(8,8),edgecolor='black')
plt.title('Team wins',fontsize=20)
plt.ylabel('count',fontsize=20)
plt.xlabel('IPL Team',fontsize=20)
```

```python
# Visualization 2
# Top 10 players won maximum man of the match award.
# Annotation
match['player_of_match'].value_counts()[0:10].plot(kind="barh",figsize=(8,8),color='g',edgecolor='brown')
plt.title('Players with maximum man of the match',fontsize=20)
plt.ylabel('Players',fontsize=20)
plt.xlabel('Number of times',fontsize=20)
```

```python
# Visualization 3
# Cities where maximum number of matches taken place.
match['city'].value_counts().plot(kind="bar",figsize=(8,8),edgecolor='black')
plt.title('Cities with maximum number of matches taken place',fontsize=15)
plt.ylabel('Number of times', fontsize=20)
plt.xlabel('Cities',fontsize=20)
```

```python
# Visualization 4
# Top 5 batsman with highest score.
delivery['batsman'].value_counts()[0:5].plot(kind="line",figsize=(8,8),marker='o',markersize=10)
plt.title('Batsman with maximum run',fontsize=20)
plt.ylabel('Runs',fontsize=20)
plt.xlabel('Batsman',fontsize=20)
```

```python
# Visualization 5
# Top 5 bowlers with highest wickets taker
delivery['bowler'].value_counts()[0:5].plot(kind="line",figsize=(8,8),marker='o',markersize =10)
plt.title('Bowlers with maximum wickets',fontsize=20)
plt.ylabel("Number of Wickets",fontsize=20)
plt.xlabel('Bowlers',fontsize=20)
```

```python
# Visualization 6
# Year maximum IPL 
match['season'].value_counts().plot(kind="bar",figsize=(8,8),edgecolor='black')
plt.title('Year mximum number of IPL matches',fontsize=20)
plt.ylabel('Number of times',fontsize=20)
plt.xlabel('Year',fontsize=20)
```
